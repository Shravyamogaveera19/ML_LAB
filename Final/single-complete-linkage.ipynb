```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

# Load dataset
data = pd.read_csv("datasets/iris.csv")

# Function to calculate proximity matrix
def proximity_matrix(data):
    n = data.shape[0]
    proximity_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(i+1, n):
            proximity_matrix[i, j] = np.linalg.norm(data.iloc[i] - data.iloc[j])
            proximity_matrix[j, i] = proximity_matrix[i, j]
    return proximity_matrix

# Function to plot dendrogram
def plot_dendrogram(data, method):
    data_array = data.values  # Convert to NumPy array
    linkage_matrix = linkage(data_array, method=method)
    dendrogram(linkage_matrix)
    plt.title(f'Dendrogram - {method} linkage')
    plt.xlabel('Data Points')
    plt.ylabel('Distance')
    plt.show()

# Use only numeric columns (exclude species)
numeric_data = data.iloc[:, :-1]

# Calculate the proximity matrix (first 6 rows for simplicity)
print("Proximity matrix:")
print(proximity_matrix(numeric_data.iloc[:6]))

# Plot dendrograms for first 6 rows
plot_dendrogram(numeric_data.iloc[:6], 'single')
plot_dendrogram(numeric_data.iloc[:6], 'complete')
```
