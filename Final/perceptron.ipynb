```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, lr=1, epochs=10):
        self.weights = np.zeros(input_size + 1)  # +1 for bias
        self.lr = lr
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        x_with_bias = np.insert(x, 0, 1)  # Add bias input as 1
        z = np.dot(x_with_bias, self.weights)
        return self.activation(z)

    def train(self, X, y):
        for _ in range(self.epochs):
            for xi, target in zip(X, y):
                xi_with_bias = np.insert(xi, 0, 1)  # Add bias input
                y_pred = self.predict(xi)
                error = target - y_pred
                self.weights += self.lr * error * xi_with_bias

# Training data
X = np.array([[0,0], [0,1], [1,0], [1,1]])

# ---------- AND Gate ----------
y_and = np.array([0, 0, 0, 1])
and_perceptron = Perceptron(input_size=2)
and_perceptron.train(X, y_and)

print("AND Gate Predictions:")
for x in X:
    print(f"{x} => {and_perceptron.predict(x)}")

# ---------- OR Gate ----------
y_or = np.array([0, 1, 1, 1])
or_perceptron = Perceptron(input_size=2)
or_perceptron.train(X, y_or)

print("\nOR Gate Predictions:")
for x in X:
    print(f"{x} => {or_perceptron.predict(x)}")

# ---------- Test input [1, 0] ----------
print("\nAND [1, 0]:", and_perceptron.predict(np.array([1,0])))
print("OR  [1, 0]:", or_perceptron.predict(np.array([1,0])))
```
