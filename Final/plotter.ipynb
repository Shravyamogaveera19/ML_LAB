```python
#Scatter Plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("datasets/ToyotaCorolla.csv")

x=data['KM']
y=data['Doors']
z= data['Price']

plt.scatter(x,y,c=z,cmap='jet')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("datasets/ToyotaCorolla.csv")

x=data['KM']
y=data['Doors']
z= data['Price']

ax = plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap = 'jet')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("datasets/ToyotaCorolla.csv")

x=data['KM']
y=data['Weight']
z= data['Price']

plt.tricontourf(x,y,z,levels=20,cmap='jet')
plt.colorbar(label = 'Price')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("datasets/ToyotaCorolla.csv")

sns.heatmap(data[['Weight','KM','Price','Doors']].corr(),cmap='jet',annot=True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("datasets/ToyotaCorolla.csv")

x=data['KM']
y=data['HP']
z= data['Price']

plt.boxplot([x,y,z])
plt.xticks([1,2,3],["KM","HP","Price"])
plt.show()
```
