```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the Iris dataset from Downloads (update the path if needed)
  # Change to your actual path
data = pd.read_csv("datasets/iris.csv")

# Extract features (drop the species column if it exists)
X = data.iloc[:, :-1].values  # All columns except the last one (species)

# Number of clusters
K = 3

# K-means using scikit-learn
kmeans = KMeans(n_clusters=K, random_state=0)
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

# Print results
print("K-means Labels:", labels)
print("K-means Centroids:\n", centroids)

# Plotting K-means results (using first two features for visualization)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=200)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering of Iris Dataset (from CSV)')
plt.show()
```
