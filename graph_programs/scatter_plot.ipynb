```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("ToyotaCorolla.csv")
x=data['KM']
y=data['Doors']
z=data['Price']
ax =plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='jet')
plt.show()
```
